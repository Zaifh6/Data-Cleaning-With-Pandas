#### Here we have an uncleaned data set about movies, we will try to clean it step by step.

#### Importing the pandas library

```python
import pandas as pd
```

#### Loading our movies dataset

```python
df = pd.read_csv('/content/movies.csv')
```

#### Checking our dataset



```python
df.head(10)
```

```python
df.info()
```

#### Renaming the columns properly

```python
df = df.rename(columns={'MOVIES': 'Movies',
                        'YEAR': 'Year',
                        'GENRE':'Genre',
                        'RATING': 'Rating',
                        'ONE-LINE': 'One-Line',
                        'STARS': 'Stars',
                        'VOTES': 'Votes',
                        'Runtime': 'Runtime',
                        'Gross': 'Gross'})
```

#### Dropping the duplicate & NAN values


```python
df = df.drop_duplicates()
df = df.dropna()
```

#### Cleaning all the columns

```python
df['Year'] = df['Year'].str.replace('(', '')
```

```python
df['Year'] = df['Year'].str.replace('\n', '')
```

```python
df['Year'] = df['Year'].str.replace(')', '')
```

```python
df['Year'] = df['Year'].str.replace('I', '')
```

```python
df['Stars'] = df['Stars'].str.replace('\n', '')
```

```python
df['Votes'] = df['Votes'].str.replace(',', '')
```

```python
df['Genre'] = df['Genre'].str.replace('\n', '')
```

#### Dropping the columns that we don't need

```python
df = df.drop(['One-Line', 'Stars'], axis=1)
```

#### Resetting our index

```python
df = df.reset_index()
```

```python
df = df.drop(['index'], axis=1)
df.index += 1
```
